```
import torch
import torch.nn as nn
import pandas as pd
import numpy as np
import tensorflow as tf
```

```
path="sample_data/"
```

```
!ls {path}
```

```
df_raw2x=pd.read_csv(f'{path}2x.csv',low_memory=False)
df_rawxy=pd.read_csv(f'{path}3x4y.csv',low_memory=False)
```

#F(x)=2x

```
df_raw2x.head()
```

```
x_train=torch.tensor(df_raw2x.X).float().reshape(1,-1)
y_train=torch.tensor(df_raw2x.Y).float().reshape(1,-1)
x_train.size(),y_train.size()
```

```
x_train=torch.transpose(x_train,0,1)
y_train=torch.transpose(y_train,0,1)
```

```
model=nn.Sequential(
       nn.Linear(1,1))
```

```
criterion=torch.nn.MSELoss()
optimiser=torch.optim.SGD(model.parameters(),lr=0.0001)
```

```
for epoch in range(50):
  y_pred=model(x_train)
  loss=criterion(y_pred,y_train)
  print('epoch: ',epoch,' loss: ',loss.item())
  optimiser.zero_grad()
  loss.backward()
  optimiser.step()
```

```
y_pred
```

```
y_pred=y_pred.eval()
y_train=y_train.eval()
import matplotlib.pyplot as plt
plt.plot(y_pred,y_train)
plt.ylabel("y_train")
plt.xlabel("y_pred")
plt.show()
```
